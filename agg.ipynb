```python
data = [['Dog', 'Golden Retriever', 1, 5], ['Dog', 'German Shepherd', 2, 5], ['Dog', 'Mule', 200, 1], ['Cat', 'Shirazi', 5, 2], ['Cat', 'Siamese', 3, 3], ['Cat', 'Sphynx', 7, 4]]
queries = pd.DataFrame(data, columns=['query_name', 'result', 'position', 'rating']).astype({'query_name':'object', 'result':'object', 'position':'Int64', 'rating':'Int64'})
queries
```

```python
queries['quality'] = queries['rating'] / queries['position'] + 1e-6
queries['poor_query_percentage'] = (queries.rating < 3) * 100
queries.groupby('query_name').agg({'quality':'mean','poor_query_percentage':'mean'}).round(2).reset_index()
```

**pandas.DataFrame.agg(func=None, axis=0, *args, **kwargs)** : https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.agg.html

* DataFrame에 다른 함수를 적용하는 방법. 특정 축에다가 특정 함수나 명령을 실행하여 DataFrame을 집계(정리, 합치기)하는 함수.
* Series 객체에다가도 사용할 수 있다.
* ***func***에다가 원하는 함수를 설정하는 것이다.


```python
"""
agg함수를 통해 queries DataFrame object를 정리했다. func매개변수에 각각의 column에 대해 어떤 함수를 적용하여 어떤
식으로 정리할 것인지를 알려준다. column label이 position인 column의 경우 min값을 찾는 함수인 min을 넣어주고 label이
rating인 column의 경우 전체 rating값들의 평균을 알아보자는 의미에서 평균값을 찾는 함수인 mean을 넣어주었다.
"""
queries.agg({'position':'min', 'rating':'mean'})
```

```python
import pandas as pd

def queries_stats(queries: pd.DataFrame) -> pd.DataFrame:
    queries['quality'] = queries.rating/queries.position + 1e-6 # 얘는 안 붙이면 오류나서 붙임. 2자리 round될 때 1.765이런 식일 때 1.77이 되어야 하는데 1.76 이런 식으로 되는 에러가 있다고 카더라(?)
    queries['poor_query_percentage'] = (queries.rating < 3) * 100
    return queries.groupby('query_name').agg({'quality':'mean','poor_query_percentage':'mean'}).round(2).reset_index()
```
