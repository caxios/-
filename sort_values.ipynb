pandas.DataFrame.sort_values(by, *, axis=0, ascending=True, inplace=False, kind='quicksort', na_position='last', ignore_index=False, key=None) : https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html

* DataFrame의 값들을 정렬하는 함수. inplace가 True면 반환값 없고, False면 sort된 DataFrame을 반환한다.
* ***axis***는 데이터를 정렬할 기준을 정해준다. 0혹은 'index'를 넣어주면 행을 기준으로, 1혹은 'columns'를 넣어주면 열을 기준으로 설정한다.
* ***by***는 정렬의 기준이 될 label을 지정하는 매개변수.
* ***na_position***은 Nan의 값을 first(맨앞)으로 보낼지 last(맨뒤)로 보낼지 정하는 매개변수
* ***kind***는 정렬 알고리즘을 지정하는 변수.
* ***key***는 정렬 시 정렬되는 데이터들에 적용할 함수를 넣어주는 자리다. by에 의해 지정된 label에 들어있는 데이터들에 sorting이 시작되기 전에 적용된다. 함수의 적용은 sorting작업에 활용되고, sorting이 끝나면 함수 적용이 해제되고 원래의 모양대로 데이터가 나온다.

```python
# axis를 0으로 설정하여 정렬이 위아래(행) 방향으로 진행될 될 것임을 알려주고, by를 통해 정렬의 기준이 될 label을 지정한다. 이 함수는 id를 기준으로 정렬이 되는데, ascending=True로
# 함으로써id값이 오름차순이 될 수 있도록 했다.
person.sort_values(axis=0, by='id', ascending=True)
```

```python
df = pd.DataFrame({
    'col1': ['A', 'A', 'B', np.nan, 'D', 'C'],
    'col2': [2, 1, 9, 8, 7, 4],
    'col3': [0, 1, 9, 4, 2, 3],
    'col4': ['a', 'B', 'c', 'D', 'e', 'F']
})
df
```

```python
"""
key에 람다함수를 넣어주었는데, 이 함수의 역할은 정렬이 진행되기 전에 정렬의 기준이 되는 label인 col1에 있는 데이터들을 소문자로 만들어주고, 정렬과정에서는 이 데이터들이 계속
소문자인 상태를 유지한다. 이후에 정렬이 다 진행이 되면 데이터가 원래 모습으로 돌아오고, inplace를 True로 안 했으므로 col1의 값들을 기준으로 정렬된 DataFrame이 반환된다. na_position도 따로
설정을 안 해줬으므로 정렬 기준이 되는 label에 있는 Nan값은 그냥 자연스레 맨 뒤로 밀려났다.
"""
df.sort_values(axis=0, by='col1', ascending=True, key=lambda col1: col1.str.lower())
```
