```python
import pandas as pd
import numpy as np
import random
```

https://leetcode.com/problems/find-customer-referee/?lang=pythondata

```python
data = [[1, 'Will', None], [2, 'Jane', None], [3, 'Alex', 2], [4, 'Bill', None], [5, 'Zack', 1], [6, 'Mark', 2]]
customer = pd.DataFrame(data, columns=['id', 'name', 'referee_id']).astype({'id':'Int64', 'name':'object', 'referee_id':'Int64'})
customer
```

pandas.DataFrame.fillna(value=None, *, method=None, axis=None, inplace=False, limit=None, downcast=_NoDefault.no_default) : https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.fillna.html
* DataFrame에서 Nan 값을 채워주는 역할을 한다.
* ***value***는 Nan을 어떤 값으로 채워 넣을 것인지 정하는 매개변수.
* ***method***는 결측값을 변경할 방식을 정한. bfill, backfill로 할 경우 결측값(Nan)을 바로 아래 값과 동일하게 변경, ffill로 할 경우 결측값을 바로 위 값과 동일하게 변경한다.

```python
# Nan값은 숫자가 아니기 때문에 DataFrame에서 숫자비교를 통해 필터링을 하려고 하면 애초에 비교 대상에서 제외된다. 그래서 비록 referee_id가 0이 아니더라도 걸러지지 않은 것.
customer[customer['referee_id']!=2]
```

```python
# customer의 빈값을 다 1로(2가 아닌 그냥 아무 수로) 바꿔준다.
df = customer.fillna(value=0)
df
```
