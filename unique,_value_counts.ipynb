```python
data = [['Dog', 'Golden Retriever', 1, 5], ['Dog', 'German Shepherd', 2, 5], ['Dog', 'Mule', 200, 1], ['Cat', 'Shirazi', 5, 2], ['Cat', 'Siamese', 3, 3], ['Cat', 'Sphynx', 7, 4]]
queries = pd.DataFrame(data, columns=['query_name', 'result', 'position', 'rating']).astype({'query_name':'object', 'result':'object', 'position':'Int64', 'rating':'Int64'})
queries
```

```python
queries['quality'] = queries['rating'] / queries['position'] + 1e-6
queries['poor_query_percentage'] = (queries.rating < 3) * 100
queries.groupby('query_name').agg({'quality':'mean','poor_query_percentage':'mean'}).round(2).reset_index()
```

**pandas.Series.unique()** : https://pandas.pydata.org/docs/reference/api/pandas.Series.unique.html
* Series object의 unique(중복되지 않는 유일한 값)들을 numpy array형식으로 반환한다.  

**pandas.Series.value_counts(normalize=False, sort=True, ascending=False, bins=None, dropna=True)** : https://pandas.pydata.org/docs/reference/api/pandas.Series.value_counts.html

* unique(고유한, 유일한)값들 각각이 몇 개씩 있는지 세어준다.
* ***normalize***는 해당 데이터의 개수가 전체 데이터에서 몇 개인지를 비율로 나타내어준다.
* ***bins***는 각각의 unique한 데이터가 몇 개 있는지를 나타내주긴 하는데 해당되는 범위 안에 얼만큼이 있는지를 알려준다.

```python
lst = pd.Series([1,2,3,4,5,5,5,5,23,3,4])
lst
```

```python
# Series객체에 있는 각각의 고유한 값들이 몇 개 있는지 세우준다. lst라는 Series객체에 5가 다섯 개 있다는 뜻.
lst.value_counts()
```

```python
# 고유한 데이터의 개수를 전체 데이터 개수의 비율로 나타낸다.
lst.value_counts(normalize=True)
```

```python
# 데이터가 1~5까지, 그리고 23이렇게 있다. 이걸 5개의 구간으로 나눠서 각각의 구간에 값들이 얼만큼씩 들어있는지 보여준다.
lst.value_counts(bins=5)
```

```python
import pandas as pd
# expected output과 동일하게 나오는데 틀린다. 뭐가 틀린지 알고 싶은데...
def queries_stats(queries: pd.DataFrame) -> pd.DataFrame:
    df = queries.drop(columns=['result'])
    df_query = df['query_name'].unique()
    ratio_lst = list()
    poor_percentage_lst = list()
    output = pd.DataFrame(columns=['query_name','quality','poor_query_percentage'])

    for i, query_name in enumerate(df_query):
        ratio_lst.append(round(sum((df[df['query_name']==query_name]['rating'] / df[df['query_name']==query_name]['position'] )) / 3, 2))
        poor_percentage_lst.append(round((df[(df['query_name']==query_name) & (df['rating']<3)]['rating'].value_counts() / df['query_name'].value_counts()[i])*100, 2))

    output['query_name'] = df_query
    output['quality'] = ratio_lst
    output['poor_query_percentage'] = poor_percentage_lst

    return output
```
