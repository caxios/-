 * ***freq***은 dataframe을 특정한 시계열(?) 순서 데이터가 datetime type이면 어떤 기준으로 groupby할 것인지 결정한다. 즉 시간빈도에 따라 데이터들을 groupby한다. M이면 month end frequency W면 weekly frequency 등. 더 많다.
 * ***key***정렬의 대상(target selection) column의 이름을 지정하는 매개변수.

```python
import pandas as pd
```

```python
sample = pd.read_excel("/content/drive/MyDrive/data/finance/snp_k200_msb_20180816_20240329.xlsx")
# apply를 보면 lambda x가 있다. Grouper를 통해 월별로 데이터를 묶고,
# groupby를 통해 월별로 묶인 데이터를 분리시켜놓은 것인데, 각각의 묶음들을
# x에 넣어주는 것. 그리고 iloc[-1]을 통해 말일 데이터를 솎아 내는 것.
sample.groupby(pd.Grouper(key='date', freq='M')).apply(lambda x: x.iloc[-1])
```
