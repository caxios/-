pandas.DataFrame.groupby(by=None, axis=_NoDefault.no_default, level=None, as_index=True, sort=True, group_keys=True, observed=_NoDefault.no_default, dropna=True) : https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html
* DataFrame의 값들을 그룹화하여 연산을 수행할 수 있도록 해주는 함수다. 그룹에 관한 정보를 담은 groupby object를 반환한다. DataFrame을 반환하는 게 아니라(?)
* ***by***는 어떤 것을 기준으로 group을 묶을 것인지 정하는 매개변수. 간단하게는 단지 하나의 열 label이 올 수 있고, Series나 list, 함수도 올 수 있다.
* ***axis***는 그룹화 할 축을 정하는 매개변수
* ***as_index***는 그룹화 한 내용을 index로 정할 지 여부를 정하는 매개변수
* ***group_keys***는 DataFrame의 method인 apply함수를 사용해서 얻은 결과(apply를 적용해 얻게 된 새로운 DataFrame)의 index가 groupby의 index와 서로 중복될 경우 groupby의 대상이 된 DataFrame의 기본 index로 설정해주는 매개변수.False면 인덱스 중복이 안 된다.
* ***dropna***는 groupby함수를 수행할 때 결측값을 제외할지 말지 정하는 매개변수.
* ***level***은 만약 groupby를 실행하는 DataFrame이 multiindex인 경우 특정 index(같은 말로 여러 index중 특정 index)를 뽑아내어 해당 index(level) 혹은 levels를 index로 삼는 것.

```python
# 링크 : https://wikidocs.net/160081
idx=['A','A','B','B','B','C','C','C','D','D','D','D','E','E','E']
col=['col1','col2','col3']
data = np.random.randint(0,9,(15,3))
df = pd.DataFrame(data=data, index=idx, columns=col).reset_index()
df
```

```python
df.columns
```

```python
# 좀 헷갈려서 보니까 df가 multiindex인 것은 아니다.
df.groupby(level=1)
```

```python
# 그룹화 후 추가적인 연산이나 labeling없이 단순히 groupby만 실행하면 groupby object만 반환할 뿐 아무런 DataFrame, Table을 보여주지 않는다.
df.groupby(by='index')
```

```python
# T는 행과 열을 바꾸는 짓. groupby에서 axis=1과 같다. 만약 원래 DataFrame에서 index를 기준으로 그룹화를 수행하고 싶다면 DataFrame에 T함수를 적용시켜야 되는 것이다.
df.T
```

```python
# col1은 이제 column label이 아니라 df DataFrame의 index가 되었으므로 groupby의 그룹화 기준이 될 수 없다.
df.T.groupby(by='col1')
```

```python
df.T.groupby(by=0)
```

```python
# agg나 apply등의 method도 사용가능한다. 연산수행의 결과 얻게 된 수치를 DataFrame으로 반환한다.
df.groupby('index').count()
```

```python
# groupby를 안 쓰고 했을 때
def project_employees_i(project: pd.DataFrame, employee: pd.DataFrame) -> pd.DataFrame:
    df = employee.merge(project, on='employee_id')
    project_id_vals = project['projezct_id'].unique()

    return_df = pd.DataFrame(columns=['project_id','average_years'])
    avg_years_lst = list()
    for project_id in project_id_vals:
        avg_year = round(df[df['project_id']==project_id]['experience_years'].sum() / df[df['project_id']==project_id].shape[0], 2)
        avg_years_lst.append(avg_year)

    return_df['project_id'] = project_id_vals
    return_df['average_years'] = avg_years_lst
    return return_df
```

```python
# groupby를 쓰면
def project_employees_i(project: pd.DataFrame, employee: pd.DataFrame) -> pd.DataFrame:
    df = employee.merge(project, on='employee_id')
    return df.groupby(by='project_id')['experience_years'].mean().round(2).rename('average_years').reset_index()
```
