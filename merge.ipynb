pandas.DataFrame.merge(right, how='inner', on=None, left_on=None, right_on=None, left_index=False, right_index=False, sort=False, suffixes=('_x', '_y'), copy=None, indicator=False, validate=None) : https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.merge.html

* 두 개의 DataFrame을 공통된 항목(교집합)을 기준으로 하나의 DataFrame으로 합치는 함수. 두 개가 하나로 합쳐진 DataFrame을 반환한다. 설명 중에 계속 병합의 기준이라는 말을 쓰는데, 이 말은 서로 다른 두 개의 DataFrame을 합칠 때 기준이 되는 column을 뜻하는 것이다. 즉, 서로 다른 두 개의 DataFrame을 합치려면 서로 공통된 특징을 가진 혹은 서로 공통된 종류의 데이터를 담은 column을 기준으로 합쳐야 한다는 뜻이다. 일종의 key값 역할을 수행하는 column을 지정한다는 것이 곧 기준을 지정해준다는 말이다.
* ***right***는 기존 DataFrame(left)에다가 합칠 DataFrame(right)을 지정해주는 매개변수다.
* ***on***은 column을 기준으로 두 DataFrame을 병합 시에, 기준으로 할 column의 이름이 각각의 DataFrame에서 동일하다면, 해당 column을 지정해줌으로써 그 공통된 column을 기준으로 DataFrame을 합치는 것이다.
* ***how***는 병합 시에 기준이 될 인덱스를 정하는 방식이다. left, right, outer, inner, cross가 있다. left는 기존객체(기존 DataFrame)를, right는 합칠(병합할) DataFrame을, inner은 두 DataFrame의 교집합을, cross는 행렬곱을 기준으로 합치는 것이다. 기본적으로 집합으로 생각하면 편하다. inner은 두 객체의 교집합된(공통된) column을 가진 row들만 합치는 거고, outer는 합집합으로 그냥 가릴 거 없이 row들을 걍 다 합치는 거다. left는 기존 DataFrame의 인덱스를 기준으로 삼기 때문에 right(합칠 DataFrame)에 없더라도 left를 기준으로 결국 합쳐지고, 없는 데이터는 Nan이 들어간다. right는 left의 반대.
* ***left_on, right_on***은 병합 시에 각각의 DataFrame(left와 right)에서 기준이 되는 column을 지정해주는 것이다. 이때 기준이 되는 각각의 column들은 서로 column의 이름은 다르더라도 각각의 row에 담긴 내용물은 같은 종류의 데이터이어야 한다.
* ***left_index, right_index***는 left_on, right_on의 index버전. True값을 넣어주면 해당 DataFrame의 인덱스가 병합의 기준이 된다.
* ***indicator***는 마지막 열에 _merge열을 추가하여 병합 정보를 출력한다.
* 열을 새로 만들게 될 때, right에 해당하는 값은 열 이름에 _y가 들어가고, left에 해당하는 값은 열 이름에 _x가 들어간다.


```python
df1 = pd.DataFrame({'IDX':['a','b','c','a'],'VAL':[1,2,3,4]})
df2 = pd.DataFrame({'IDX':['a','c','d'],'VAL':[5,6,7]})
print(df1)
print(df2, " \n")
"""
두 DataFrame(df1,df2)의 IDX column은 서로 공통된 column이다. on을 IDX로 지정하여 두 DataFrame이 합쳐지는 기준(공통점)이 IDX라고 명시적으로 밝혀줬고,
how를 left(df1)로 지정함으로써 df1의 IDX column에 있는 값들이 병합의 기준이 된다. 두 번째 DataFrame인 df2의 경우 IDX가 b인 row가 없는데, 때문에 병합 후에
b row의 VAL_y가 Nan이다.
"""
print("merge함수 사용한 결과\n------------------->")
print(df1.merge(df2,how='left',on='IDX')) #=print(pd.merge(df1,df2, how='left',on='IDX'))
```

```python
# 두 DataFrame의 column을 확인해보면 VAL의 값들이 서로 겹치는 게 없다. 그러므로 left의 VAL을 기준으로 합쳤을 때, 각각 1,2,3,4 row의 IDX_y가 Nan인 것이다.
print(df1.merge(df2,how='left',on='VAL'))
```

```python
df3 = pd.DataFrame({'IDX1':['a','b','c','a'],'VAL':[1,2,3,4]})
df4 = pd.DataFrame({'IDX2':['a','c','d'],'VAL':[5,6,7]})
print(df3)
print(df4)
"""
df3의 IDX1과 df4의 IDX2는 서로 column의 이름이 다르지만, 각 column에 들어있는 데이터들의 종류를 보면 a,b,c..로 알파벳이라는 공통된 특징이 있다.
따라서 두 DataFrame의 공통된 부분 즉, 병합의 기준이 되는 column이 이름은 다르지만 데이터는 같은 종류이므로 이를 기준으로 병합을 하는 것이다. how에 left(df3)를
넣어줌으로써 이번에도 역시 left(df3) DataFrame의 IDX1에 있는 값들이 병합의 기준이 된다. 각 DataFrame의 두 column이름이 공통되지 않으므로 on이 아니라
left_on과 right_on을 써주는데, left와 right의 column을 살펴봤을 때 서로 데이터의 종류가 겹치는 부분이 각각 IDX1과 IDX2이므로 이를 지정해줘야 한다.
"""
print("merge함수 사용한 결과\n------------------->")
print(df3.merge(df4, how='left', left_on='IDX1', right_on='IDX2'))
```

```python
"""
how에 left를 넣어줌으로써 left(employee)를 기준으로 DataFrame을 병합한다. 기준으로 삼은 값은 managerId. 그래서 left_on에 managerId가 들어간 것.
right 즉, 합쳐지게 될 DataFrame(역시나 employee) 부분 중 managerId와 겹치는 부분은 id와 managerId인데 이중 id를 선택한 이유는 리트코드 문제 해결 시 그렇게 하는
편이 낫기 때문.
"""
emp = employee.merge(employee, how='left', left_on='managerId', right_on='id')
emp
```
