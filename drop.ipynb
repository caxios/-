pandas.DataFrame.drop(labels=None, *, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise') : https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html

* DataFrame에서 삭제(drop)하고 싶은 녀석을 drop한 다음, drop된 DataFrame을 반환하는 함수.
* ***labels***는 삭제할 column label을 지정하는 매개변수다. axis가 기본이 0(index)이므로 column label로 자르려면 axis를 1로 넣고 해야한다. axis를 1로 설정하지 않으면 그냥 여기서의 label은 index label을 의미하게 된다.
* ***axis***는 0이면 index기준으로 drop한다는 뜻이고, 1이면 column기준으로 drop한다는 뜻이다.
* ***index***는 index를 기준으로 row를 빼고 싶으면 해당하는 index값들을 넣어주면 된다.
* ***errors***는 삭제할 label을 찾지 못했을 때 error를 띄울지 말지를 결정하는 녀석이다.

```python
df = pd.DataFrame(np.arange(12).reshape(3, 4),
                  columns=['A', 'B', 'C', 'D'])
df
```

```python
# 이런 식이면 column을 drop한다. label이 B, C인 column이 떨어졌다.
df.drop(['B', 'C'], axis=1)
```

```python
# 숫자로 넣어주면 row 단위로 떨굴 수 있다. 숫자들은 index다.
df.drop([0, 1])
```

```python
# 이건 prefix에 있는 특수기호들을 제대로 검사하지 못해 답은 아니다. 걍 일단 직관적인 풀이.
#abc_lst = 'a b c d e f g h i j k l m n o p q r s t u v w x y z _ . -'.split(' ')
for i, mail in enumerate(users['mail']):
  print(f'now is {i}th {mail} checking')
  if '@leetcode.com' not in mail:
    print(f"Drop {mail} since no @leetcode.com")
    users = users.drop([i])
  elif (mail[0] in ['_','.','-']) or (mail[0] in '1 2 3 4 5 6 7 8 9 0'.split(' ')):
    print(f'Drop {mail} since invalid starter')
    users = users.drop([i])
  prefix = mail.split('@leetcode.com')[0]
  if ('#' in [letter for letter in prefix]):
    print("I found #!!!!!!")
    users = users.drop([i])
```

```python
# regex(정규표현식)를 활용한 방식. 사실 가장 직관적이고 간단한 풀이긴 하다..
def valid_emails(users: pd.DataFrame) -> pd.DataFrame:
    df = users[users['mail'].str.match(r'^[A-Za-z][A-Za-z0-9_\.\-]*@leetcode(\?com)?\.com$')]

    return df
```
